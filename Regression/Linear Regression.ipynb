```python
from scipy import stats
corr_df = pd.DataFrame(columns=['p'])

for col in df:
    print(col)
    if pd.api.types.is_numeric_dtype(df[col]):
        r,p = stats.pearsonr(df.MV,df[col])
        corr_df.loc[col]=[(p)]
corr_df
```
