```python
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"E:\Data Science\Assignment-4\boston.xls")
print(df)
```

```python
x = df[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PT','B','LSTAT']]
y = df[['MV']]
```

```python
x.shape
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3, n_init = 3, init = "random", random_state = 42)
km.fit(x)
y_kmeans = km.predict(x)
y_kmeans
```

```python
import seaborn as sns
sns.scatterplot(x,palette = "coolwarm_r")
centers = km.cluster_centers_

plt.scatter(centers[:, 0], centers[:, 1], c='black', s=100, alpha = 0.3);
plt.xlabel("X-point")
plt.ylabel("Decision")
```

```python
intertia = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(x)
    intertia.append(km.inertia_)
plt.plot(K, intertia, marker= "x")
plt.xlabel('k')
plt.xticks(np.arange(15))
plt.ylabel('Intertia')
plt.title('Elbow Method')
plt.show()
```
